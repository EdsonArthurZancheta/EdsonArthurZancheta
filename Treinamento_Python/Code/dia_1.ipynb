## Dicionário de Dados
 - **survival**: Indica se o passageiro sobreviveu ou não (0 = Não, 1 = Sim).
 - **pclass**: Classe do ticket do passageiro (1 = 1ª classe, 2 = 2ª classe, 3 = 3ª classe).
 - **sex**: Sexo do passageiro.
 - **age**: Idade do passageiro em anos (a idade é fracionária se menor que 1; se estimada, é no formato xx.5).
 - **sibsp**: Número de irmãos/cônjuges a bordo do Titanic.
 - **parch**: Número de pais/filhos a bordo do Titanic.
 - **ticket**: Número do ticket do passageiro.
 - **fare**: Tarifa paga pelo passageiro.
 - **cabin**: Número da cabine do passageiro.
 - **embarked**: Porto de embarque (C = Cherbourg, Q = Queenstown, S = Southampton).

### Notas sobre as Variáveis
 - **pclass**: Uma proxy para o status socioeconômico (SES) do passageiro:
   - 1ª classe = Alta
   - 2ª classe = Média
   - 3ª classe = Baixa
 - **age**: A idade é fracionária se menor que 1. Se a idade é estimada, está no formato xx.5.
 - **sibsp**:  Quantidade de familiares presentes no dataset. Esposas e Irmãos
 - **parch**: Quantidade de Pai, mãe ou filho presente no dataset. 

Portanto se determinada pessoa levou 1 irmão e sua esposa **sibsp = 2** <br>
Algumas crianças viajaram apenas com uma babá, portanto **parch = 0** para elas.

#### 6.2 Aplicação de Funções com .apply() e Funções Lambda <br>
Podemos aplicar transformações personalizadas nos dados usando `.apply()` e funções `lambda`. Esse método é útil para automatizar cálculos ou criar novas colunas baseadas em condições.

Exemplo de Função Lambda
Vamos criar uma coluna para categorizar a idade dos passageiros como **"Menor de idade" e "Maior de idade".**

```python
# Como funciona uma função lambda e apply?
# Vamos relembrar o conceito de loop for e if.

idades = [18, 49, 26, 45, 56, 52, 32, 40]

# Começando com o loop for que irá percorrer cada idade da lista idades
for idade in idades:
    print(idade)
```

```python

# Agora criando o if else para verificar se a idade é maior ou menor que 18
if idade < 18:
    print('Menor de idade')
else: 
    print('Maior de idade')
```

```python
# se juntarmos ambas as estruturas, teremos um loop for com if else:

for idade in idades:
    if idade < 18:
        print('Menor de idade')
    else: 
        print('Maior de idade')
```

 ##### Função lambda para classificar idade
 
 Note que `.apply()` e `lambda` são distintos. <br>
 Apply é uma função que aplica uma função a cada elemento de uma série ou dataframe,
 é como se o apply substituisse o loop for.

 Lambda é uma função anônima que pode ser usada para criar funções simples.

```python
data['Age'].apply(lambda x: 'Menor de idade' if x < 18 else 'Maior de idade')

# x equivale a cada item da lista de idades
# se x for menor que 18, a função lambda irá retornar 'Menor de idade'
# caso contrário, irá retornar 'Maior de idade'
```

Outro modo de utilizar o apply seria transformarmos o nosso código em uma função declarada para facilitar legibilidade. porém isto é recomendado apenas para casos que iremos reutilizar a função em outros momentos do código.


```python
def classificar_idade(idade):
    if idade < 18:
        return 'Menor de idade'
    else:
        return 'Maior de idade'
    
# Funciona da mesma forma que a função lambda
data['Age'].apply(classificar_idade)

```

```python
# Agora aplicamos a tratativa em nosso dataframe
# Em uma nova coluna chamada 
data['AgeCategory'] = data['Age'].apply(classificar_idade)
data[['Age', 'AgeCategory']].head()
```

Agora podemos avaliar se menores de idade foram priorizados no resgate:

```python
data.pivot_table(values='Survived', index='AgeCategory', columns='Sex', aggfunc='sum')
```

E se eu desejar saber a proporção de sobreviventes?

A proporção pode ser calculada como:

Qtd de passageiros sobreviventes / total de passageiros.

Para saber quais são os sobreviventes podemos somar o campo survived.
Para saber qual o total de passageiros podemos realizar um count.

```python
# Passo 1: Contagem total de passageiros por AgeCategory e Sex
total_passageiros = data.pivot_table(values='Survived', index='AgeCategory', columns='Sex', aggfunc='count')

# Passo 2: Soma dos sobreviventes por AgeCategory e Sex
passageiros_sobreviventes = data.pivot_table(values='Survived', index='AgeCategory', columns='Sex', aggfunc='sum')

# Passo 3: Proporção de sobreviventes
proporcao_sobrevivencia = passageiros_sobreviventes / total_passageiros
proporcao_sobrevivencia
```

#### 6.3 Manipulação de Dados Categóricos com One-Hot Encoding
Para algoritmos de machine learning, é comum transformar variáveis categóricas em variáveis numéricas. O método `pd.get_dummies()` realiza essa transformação, criando uma coluna para cada valor único, com valores binários (0 - False ou 1 - True).

<q>Dados numéricos são preferíveis</q>




```python

# Codificação One-Hot para a coluna 'Pclass', mantendo a coluna original
pclass_dummies = pd.get_dummies(data['Pclass'], prefix='Class')
pclass_dummies
```

```python
data = pd.concat([data,
                  pclass_dummies],
                  axis=1)

# Exibe as primeiras linhas para verificar que 'Pclass' está presente junto com as dummies
data[['Pclass', 'Class_1', 'Class_2', 'Class_3']].head()


# !!! IMPORTANTE !!!
# É recomendado realizar isto após a divisão do dataset em treino e teste.
# Porém iremos aplicar agora para fins didáticos.
# Se tiver interesse em entender mais, pesquise sobre "Data Leakage" em ciência de dados.

# Em resumo o correto é aplicar o One-Hot Encoding após a divisão do dataset em treino e teste (que veremos no próximo dia)
# utilizando a biblioteca do scikit-learn. Porém isto gera certa complexidade para quem está iniciando.
```

#### Por que Sex e AgeCategory não são necessários no one hot encoding?
One hot encoding é recomendado para cenários de variáveis não binárias. 
Para casos binários, podemos aplicar uma tratativa mais simples apenas transformando os valores em 0 e 1.

```python
# Mapeamento binário para a coluna 'Sex'
data['Sex_bin'] = data['Sex'].map({'male': 0, 
                                   'female': 1})

# Mapeamento binário para a coluna 'Age'
data['AgeCategory_bin'] = data['AgeCategory'].map({'Maior de idade': 1, 
                                                   'Menor de idade': 0})
```

- Os "bigodes" do gráfico (linhas que se estendem da caixa) mostram a extensão dos dados que não são considerados outliers, ou seja, que estão dentro de 1,5 vezes o intervalo interquartil.

- Pontos fora dos bigodes indicam possíveis outliers, que são valores significativamente diferentes da maioria dos dados.

```python

# 7. Gráfico de violino para 'Age' por 'Pclass'
# Este gráfico mostra a densidade da distribuição da idade em cada classe. Pode ajudar a entender a diferença de idade entre as classes e o perfil dos passageiros.
plt.figure(figsize=(10, 6))
sns.violinplot(x='Pclass', y='Age', data=data, palette='viridis')
plt.title("Distribuição da Idade por Classe")
plt.xlabel("Classe")
plt.ylabel("Idade")
plt.show()
```

Na interpretação de um violin plot:

- A espessura da "viola" mostra a densidade dos dados em cada faixa do eixo y. Quanto mais larga a área em uma região, mais valores estão concentrados ali.

- Ele também pode conter uma linha interna similar ao box plot, com a mediana e os quartis destacados, mostrando o resumo estatístico dos dados.

- O gráfico permite perceber se a distribuição é unimodal (um pico), bimodal (dois picos), simétrica ou assimétrica.

```python
# 1. Gráfico de contagem para 'Survived' por 'Pclass' (variável categórica x categórica)
# Este gráfico mostra a distribuição de sobrevivência em cada classe.
plt.figure(figsize=(8, 6))

sns.countplot(x='Pclass', hue='Survived', data=data, palette='viridis')

plt.title("Sobrevivência por Classe")
plt.xlabel("Classe")
plt.ylabel("Contagem")

# Nova função para adicionar legenda ao gráfico
plt.legend(title="Sobreviveu", loc="upper right", labels=["Não", "Sim"])

plt.show()
```

```python
# 2. Gráfico de contagem para 'Survived' por 'Sex' (variável categórica x categórica)
# Este gráfico mostra a sobrevivência em relação ao sexo, destacando se a proporção de sobreviventes entre homens e mulheres foi diferente.
plt.figure(figsize=(8, 6))
sns.countplot(x='Sex', hue='Survived', data=data, palette='viridis')
plt.title("Sobrevivência por Sexo")
plt.xlabel("Sexo")
plt.ylabel("Contagem")
plt.legend(title="Sobreviveu", loc="upper right", labels=["Não", "Sim"])
plt.show()
```

```python
# 3. Gráfico de dispersão (scatterplot) para 'Age' vs 'Fare' com a distinção de 'Survived' (variável contínua x contínua)
# Um gráfico de dispersão para observar a relação entre idade e tarifa, diferenciando pela sobrevivência.
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='Fare', hue='Survived', data=data, palette='viridis')
plt.title("Idade vs Tarifa (Diferenciação por Sobrevivência)")
plt.xlabel("Idade")
plt.ylabel("Tarifa")
plt.legend(title="Sobreviveu", loc="upper right")
plt.show()
```

```python
# 4. Gráfico de violino para 'Age' por 'Survived' com separação por 'Sex' (variável contínua x categórica x categórica)
# Este gráfico exibe a distribuição de idade entre sobreviventes e não sobreviventes, diferenciados por sexo.
plt.figure(figsize=(10, 6))
sns.violinplot(x='Survived', y='Age', hue='Sex', data=data, split=True, palette='viridis')
plt.title("Distribuição de Idade por Sobrevivência e Sexo")
plt.xlabel("Sobreviveu")
plt.ylabel("Idade")
plt.legend(title="Sexo", loc="upper right")
plt.show()
```

```python
# 5. Gráfico de barra para 'Pclass' por 'Survived' com a tarifa média (variável categórica x categórica x contínua)
# A tarifa média para cada classe é comparada entre sobreviventes e não sobreviventes.
plt.figure(figsize=(10, 6))

sns.barplot(x='Pclass', y='Fare', hue='Survived', data=data, palette='viridis', errorbar=None)

plt.title("Tarifa Média por Classe e Sobrevivência")
plt.xlabel("Classe")
plt.ylabel("Tarifa Média")
plt.legend(title="Sobreviveu", loc="upper right", labels=["Não", "Sim"])
plt.show()
```

```python
# 6. Gráfico de calor (heatmap) para a correlação entre variáveis numéricas (continua x contínua)
# Um mapa de calor mostra as correlações entre variáveis como idade, tarifa e sobrevivência.
plt.figure(figsize=(8, 6))
correlation_matrix = data[['Survived', 'Pclass', 'Age', 'Fare', 'SibSp', 'Parch']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='viridis')
plt.title("Mapa de Calor das Correlações entre Variáveis Numéricas")
plt.show()
```
